```python
# Import Dependencies
import pandas as pd
import gmaps
import requests
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

# Google developer API key
from config import google_key

# Access maps with unique API key
gmaps.configure(api_key=google_key)
```

```python
#read excel files
dispensary_df = pd.read_csv("CO Recreational Marijuana Stores.csv")

cities = dispensary_df['City ']
len(cities)
```

```python
lat = []
lng = []
for city in cities:
    target_city = f'{city}, Colorado'
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(target_city, google_key)
    geo_data = requests.get(target_url).json()
    lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
    lng.append(geo_data["results"][0]["geometry"]["location"]["lng"])
    print(f'Retrieving data for {city}')
    print('---------------------------')
```

```python
lat = list(map(float, lat))
lng = list(map(float, lng))
```

```python
# make lat and lng into df

coords_df = pd.DataFrame({
    
    "City": cities,
    "LAT": lat,
    "LNG": lng,
    
})

coordinates = coords_df[["LAT", "LNG"]]
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '900px',
    'height': '900px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python

markers = gmaps.marker_layer(coordinates)
fig.add_layer(markers)
```

```python
coords_df['Coordinates'] = list(zip(coords_df.LNG, coords_df.LAT))
```

```python
coords_df['Coordinates'] = coords_df['Coordinates'].apply(Point)
```

```python
geo_df = gpd.GeoDataFrame(coords_df, geometry='Coordinates')
```

```python
#read excel files
dispensary_df = pd.read_csv("CO Recreational Marijuana Stores.csv")
colorado_df = pd.read_csv("Income/semi_cleaned_income.csv")

license = dispensary_df['License #']
street = dispensary_df['Street']
cities = dispensary_df['City']
zip = dispensary_df['Zip']

dispensary_df = pd.DataFrame({
    
    "License": license,
    "Street": street,
    "City": cities,
    "Zipcode": zip,
})
```

```python
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged.head()

merged.to_csv("Income/income_merged_with_dispensary.csv", index=False)
```

```python
colorado_df = pd.read_csv("Employment/semi_cleaned_employment_2012.csv")
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged = merged.dropna()
merged.to_csv("Employment/employment_2012_with_dispensary.csv", index=False)

colorado_df = pd.read_csv("Employment/semi_cleaned_employment_2013.csv")
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged = merged.dropna()
merged.to_csv("Employment/employment_2013_with_dispensary.csv", index=False)

colorado_df = pd.read_csv("Employment/semi_cleaned_employment_2014.csv")
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged = merged.dropna()
merged.to_csv("Employment/employment_2014_with_dispensary.csv", index=False)

colorado_df = pd.read_csv("Employment/semi_cleaned_employment_2015.csv")
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged = merged.dropna()
merged.to_csv("Employment/employment_2015_with_dispensary.csv", index=False)

colorado_df = pd.read_csv("Employment/semi_cleaned_employment_2016.csv")
merged = pd.merge(colorado_df, dispensary_df, on='Zipcode', how='left')
merged = merged.dropna()
merged.to_csv("Employment/employment_2016_with_dispensary.csv", index=False)
```

```python
employment_2012_df = pd.read_csv("Employment/semi_cleaned_employment_2012.csv")
employment_2012_df = employment_2012_df.dropna()

employment_2013_df = pd.read_csv("Employment/employment_2013_with_dispensary.csv")
employment_2013_df = employment_2013_df.dropna()

employment_2014_df = pd.read_csv("Employment/employment_2014_with_dispensary.csv")
employment_2014_df = employment_2014_df.dropna()

employment_2015_df = pd.read_csv("Employment/employment_2015_with_dispensary.csv")
employment_2015_df = employment_2015_df.dropna()

employment_2016_df = pd.read_csv("Employment/employment_2016_with_dispensary.csv")
employment_2016_df = employment_2016_df.dropna()
```

```python
avg_total_2012 = employment_2012_df['Total Unemployment 2012'].mean()
avg_total_2013 = employment_2013_df['Total Unemployment 2013'].mean()
avg_total_2014 = employment_2014_df['Total Unemployment 2014'].mean()
avg_total_2015 = employment_2015_df['Total Unemployment 2015'].mean()
avg_total_2016 = employment_2016_df['Total Unemployment 2016'].mean()


avg_total_sixteen_2012 = employment_2012_df['Age Group 16-19 2012 Unemployment'].mean()
avg_total_twenty_2012 = employment_2012_df['Age Group 20-24 2012 Unemployment'].mean()
avg_total_twenty_five_2012 = employment_2012_df['Age Group 25-44 2012 Unemployment'].mean()
avg_total_forty_five_2012 = employment_2012_df['Age Group 45-54 2012 Unemployment'].mean()
avg_total_fifty_five_2012 = employment_2012_df['Age Group 55-64 2012 Unemployment'].mean()
avg_total_sixty_five_2012 = employment_2012_df['Age Group 65-74 2012 Unemployment'].mean()
avg_total_seventy_five_2012 = employment_2012_df['Age Group 75+ 2012 Unemployment'].mean()

avg_total_sixteen_2013 = employment_2013_df['Age Group 16-19 2013 Unemployment'].mean()
avg_total_twenty_2013 = employment_2013_df['Age Group 20-24 2013 Unemployment'].mean()
avg_total_twenty_five_2013 = employment_2013_df['Age Group 25-44 2013 Unemployment'].mean()
avg_total_forty_five_2013 = employment_2013_df['Age Group 45-54 2013 Unemployment'].mean()
avg_total_fifty_five_2013 = employment_2013_df['Age Group 55-64 2013 Unemployment'].mean()
avg_total_sixty_five_2013 = employment_2013_df['Age Group 65-74 2013 Unemployment'].mean()
avg_total_seventy_five_2013 = employment_2013_df['Age Group 75+ 2013 Unemployment'].mean()

avg_total_sixteen_2014 = employment_2014_df['Age Group 16-19 2014 Unemployment'].mean()
avg_total_twenty_2014 = employment_2014_df['Age Group 20-24 2014 Unemployment'].mean()
avg_total_twenty_five_2014 = employment_2014_df['Age Group 25-44 2014 Unemployment'].mean()
avg_total_forty_five_2014 = employment_2014_df['Age Group 45-54 2014 Unemployment'].mean()
avg_total_fifty_five_2014 = employment_2014_df['Age Group 55-64 2014 Unemployment'].mean()
avg_total_sixty_five_2014 = employment_2014_df['Age Group 65-74 2014 Unemployment'].mean()
avg_total_seventy_five_2014 = employment_2014_df['Age Group 75+ 2014 Unemployment'].mean()

avg_total_sixteen_2015 = employment_2015_df['Age Group 16-19 2015 Unemployment'].mean()
avg_total_twenty_2015 = employment_2015_df['Age Group 20-24 2015 Unemployment'].mean()
avg_total_twenty_five_2015 = employment_2015_df['Age Group 25-29 2015 Unemployment'].mean()
avg_total_thirty_2015 = employment_2015_df['Age Group 30-34 2015 Unemployment'].mean()
avg_total_thirty_five_2015 = employment_2015_df['Age Group 35-44 2015 Unemployment'].mean()
avg_total_25_44_2015 = (avg_total_twenty_five_2015 + avg_total_thirty_2015 + avg_total_thirty_five_2015) / 3
avg_total_forty_five_2015 = employment_2015_df['Age Group 45-54 2015 Unemployment'].mean()
avg_total_fifty_five_2015 = employment_2015_df['Age Group 55-64 2015 Unemployment'].mean()
avg_total_sixty_five_2015 = employment_2015_df['Age Group 65-74 2015 Unemployment'].mean()
avg_total_seventy_five_2015 = employment_2015_df['Age Group 75+ 2015 Unemployment'].mean()

avg_total_sixteen_2016 = employment_2016_df['Age Group 16-19 2016 Unemployment'].mean()
avg_total_twenty_2016 = employment_2016_df['Age Group 20-24 2016 Unemployment'].mean()
avg_total_twenty_five_2016 = employment_2016_df['Age Group 25-29 2016 Unemployment'].mean()
avg_total_thirty_2016 = employment_2016_df['Age Group 30-34 2016 Unemployment'].mean()
avg_total_thirty_five_2016 = employment_2016_df['Age Group 35-44 2016 Unemployment'].mean()
avg_total_25_44_2016 = (avg_total_twenty_five_2016 + avg_total_thirty_2016 + avg_total_thirty_five_2016) / 3
avg_total_forty_five_2016 = employment_2016_df['Age Group 45-54 2016 Unemployment'].mean()
avg_total_fifty_five_2016 = employment_2016_df['Age Group 55-64 2016 Unemployment'].mean()
avg_total_sixty_five_2016 = employment_2016_df['Age Group 65-74 2016 Unemployment'].mean()
avg_total_seventy_five_2016 = employment_2016_df['Age Group 75+ 2016 Unemployment'].mean()

x_axis = ["2012", "2013", "2014", "2015", "2016"]
y_axis_total = [avg_total_2012, avg_total_2013, avg_total_2014, avg_total_2015, avg_total_2016]

y_axis_sixteen = [avg_total_sixteen_2012, avg_total_sixteen_2013, avg_total_sixteen_2014, avg_total_sixteen_2015, avg_total_sixteen_2016]
y_axis_twenty = [avg_total_twenty_2012, avg_total_twenty_2013, avg_total_twenty_2014, avg_total_twenty_2015, avg_total_twenty_2016]
y_axis_twenty_five = [avg_total_twenty_five_2012, avg_total_twenty_five_2013, avg_total_twenty_five_2014, avg_total_25_44_2015, avg_total_25_44_2016]
y_axis_forty_five = [avg_total_forty_five_2012, avg_total_forty_five_2013, avg_total_forty_five_2014, avg_total_forty_five_2015, avg_total_forty_five_2016]
y_axis_fifty_five = [avg_total_fifty_five_2012, avg_total_fifty_five_2013, avg_total_fifty_five_2014, avg_total_fifty_five_2015, avg_total_fifty_five_2016]
y_axis_sixty_five = [avg_total_sixty_five_2012, avg_total_sixty_five_2013, avg_total_sixty_five_2014, avg_total_sixty_five_2015, avg_total_sixty_five_2016]
y_axis_seventy_five = [avg_total_seventy_five_2012, avg_total_seventy_five_2013, avg_total_seventy_five_2014, avg_total_seventy_five_2015, avg_total_seventy_five_2016]

plt.plot(x_axis, y_axis_total, color='red')
plt.plot(x_axis, y_axis_twenty)
plt.plot(x_axis, y_axis_twenty_five)
plt.plot(x_axis, y_axis_forty_five)
plt.plot(x_axis, y_axis_fifty_five)
plt.plot(x_axis, y_axis_sixty_five)
plt.plot(x_axis, y_axis_seventy_five)
plt.xlabel("Years")
plt.ylabel("Unemployment (%)")
plt.title("Overall Colorado Unemployment for Cities with Dispnesaries")
plt.gcf().set_size_inches((15, 10))  
plt.grid(True)
plt.savefig("/Overall Colorado Unemployment for Cities with Dispnesaries.png")
plt.show()
```

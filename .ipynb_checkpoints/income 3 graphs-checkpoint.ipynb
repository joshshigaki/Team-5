```python
cities = dispensary_df['City']

lat = []
lng = []
for city in cities:
    target_city = f'{city}, Colorado'
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(target_city, google_key)
    geo_data = requests.get(target_url).json()
    lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
    lng.append(geo_data["results"][0]["geometry"]["location"]["lng"])
    print(f'Retrieving data for {city}')
    print('---------------------------')
```

```python
lat = list(map(float, lat))
lng = list(map(float, lng))
```

```python
# make lat and lng into df

coords_df = pd.DataFrame({
    
    "City": cities,
    "LAT": lat,
    "LNG": lng,
    
})

coordinates = coords_df[["LAT", "LNG"]]
```
